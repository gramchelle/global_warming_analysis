The term "global warming" describes the steady, long-term rise in Earth's atmosphere's average temperature brought on by the greenhouse effect. Gases from many human activities, but primarily from transportation, are responsible for this effect. We refer to the gasses as CO2 emissions.

According to the paragraph, the countries who has the highest CO2 emission shall be the most affected by global warming.
## Global Warming Data

We have created a single dataset to use. A summary and preview are provided below. It is the original data itself.

| Column     | Description              |
|------------|--------------------------|
| `'countries'` | Names of the countries. |
| `'carbon_emissions'` | Carbon emission in tons per capita. |
| `'global_warming_cope'` | Shows if the country is affected by global warming or not. |
| `'gdp'` | Shows the gross domestic product. |
| `'pop_density_growth'` |  |
| `'pop_density'` | Ratio of people per square kilometer. |
| `'region'` | Names of the regions for the countries. |

## India and Finland Records Data

A set of data collected from the external resources.

| Column     | Description              |
|------------|--------------------------|
| `'year'` | Shows the years. |
| `'forest_cover'` | Shows the area covered by forest in million metrics. |
| `'ratio_percentage'` | Shows the forest_cover ratio per km square. |
| `'population'` | Shows the population of the country in that year. |
| `'co2_emissions'` | Shows the CO2 emissions of the country in that year. |

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

india_records = pd.read_excel("india_records.xlsx")

sns.regplot(x='population', y='co2_emissions', data=india_records, scatter_kws={'s':50}, line_kws={'color':'blue'})

correlation = india_records['population'].corr(india_records['co2_emissions'])
print(f'Correlation coefficient: {correlation}')

plt.xlabel("Population Density (in Millions)")
plt.ylabel("CO2 Emission (in Millions)")
plt.title('Population vs CO2 Emission for India')
plt.show()

india_records.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

india_records = pd.read_excel("india_records.xlsx")

sns.regplot(x='ratio_percentage', y='co2_emissions', data=india_records, scatter_kws={'s':50}, line_kws={'color':'blue'})

correlation = india_records['ratio_percentage'].corr(india_records['co2_emissions'])
print(f'Correlation coefficient: {correlation}')

plt.xlabel("Forest Cover Percentage Per Km")
plt.ylabel("CO2 Emission (in Millions)")
plt.title('Forest Coverage Percentage vs CO2 Emission for India')
plt.show()

india_records.head()
```
